```python
import nltk
nltk.download('punkt')

from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer

marathi_text = """
पुणे, महाराष्ट्राची सांस्कृतिक राजधानी, शिक्षण आणि उद्योगाचे एक प्रमुख केंद्र आहे.
शहराला एक श्रीमंत ऐतिहासिक पार्श्वभूमी आहे, जिथे मराठा साम्राज्याचे मूळ स्थान होते.
आज, पुणे हे माहिती तंत्रज्ञान आणि ऑटोमोबाइल उद्योगाचे एक मोठे केंद्र बनले आहे.
येथील हवामान आल्हाददायक असते आणि अनेक पर्यटन स्थळे आहेत.
"""


parser = PlaintextParser.from_string(marathi_text, Tokenizer("english"))


summarizer = TextRankSummarizer()

summary = summarizer(parser.document, 2)

print("--- TextRank (Sumy) वर आधारित सारांश ---")
for sentence in summary:
    print(sentence)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "ai4bharat/IndicBART-XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

marathi_text = """
पुणे, महाराष्ट्राची सांस्कृतिक राजधानी, शिक्षण आणि उद्योगाचे एक प्रमुख केंद्र आहे.
शहराला एक श्रीमंत ऐतिहासिक पार्श्वभूमी आहे, जिथे मराठा साम्राज्याचे मूळ स्थान होते.
आज, पुणे हे माहिती तंत्रज्ञान आणि ऑटोमोबाइल उद्योgaचे एक मोठे केंद्र बनले आहे.
येथील हवामान आल्हाददायक असते आणि अनेक पर्यटन स्थळे आहेत.
शिक्षण क्षेत्रातही पुणे विद्यापीठाचे नाव जगभरात प्रसिद्ध आहे.
"""

input_text = f"{marathi_text} </s> <2mr>"
inputs = tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)

summary_ids = model.generate(
    inputs.input_ids,
    num_beams=4,
    min_length=30,
    max_length=100,
    length_penalty=2.0,
    early_stopping=True,
    decoder_start_token_id=tokenizer.encode("<2mr>", add_special_tokens=False)[0]
)

summary = tokenizer.batch_decode(
    summary_ids,
    skip_special_tokens=True,
    clean_up_tokenization_spaces=False
)[0]

print("--- ॲबस्ट्रॅक्टिव सारांश (लांब) ---")
print(summary)
```

```python
!pip install transformers sentencepiece --quiet

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

print("मॉडेल लोड होत आहे... (Loading model, please wait...)")

model_name = "ai4bharat/IndicBART-XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)

print("मॉडेल लोड झाले. (Model loaded.)")

def summarize_marathi(text_to_summarize):
    """
    या फंक्शनला मराठी मजकूर दिल्यास, ते त्याचा सारांश परत करते.
    """

    input_text = f"{text_to_summarize} </s> <2mr>"


    inputs = tokenizer(
        input_text,
        return_tensors="pt",
        max_length=1024,
        truncation=True
    ).to(device)


    summary_ids = model.generate(
        inputs.input_ids,
        num_beams=4,
        min_length=30,   # Your parameter
        max_length=100,  # Your parameter
        length_penalty=2.0, # Your parameter
        early_stopping=True,
        decoder_start_token_id=tokenizer.encode("<2mr>", add_special_tokens=False)[0]
    )


    summary = tokenizer.batch_decode(
        summary_ids,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=False
    )[0]

    return summary



user_text = input("कृपया तुमचा मराठी मजकूर येथे टाका (Please paste your Marathi text here):\n")

if user_text:

    final_summary = summarize_marathi(user_text)

    print("\n--- तुमचा ॲबस्ट्रॅक्टिव सारांश ---")
    print(final_summary)
else:
    print("तुम्ही कोणताही मजकूर टाकला नाही.")
```
